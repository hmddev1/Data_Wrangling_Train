### Point:
As you can see, there is no name for columns in dataframe. 
We have to ask from data file creator about what are those. Then put the names for all of columns.
After listing names, we define the name's list then use <b>read_csv()</b> to import them to the data frame.

```python
headers = ["symboling","normalized-losses","make","fuel-type","aspiration", "num-of-doors","body-style",
         "drive-wheels","engine-location","wheel-base", "length","width","height","curb-weight","engine-type",
         "num-of-cylinders", "engine-size","fuel-system","bore","stroke","compression-ratio","horsepower",
         "peak-rpm","city-mpg","highway-mpg","price"]
```

```python
df = pd.read_csv("imports-85.data", names = headers)
df.head(5)
```

### Point:
There is a lots of missing data as <b>?</b>. So we must raplace them with <b>NaN (Not a Number)</b> which Pandas knows them as missing data.
* We do this with Numpy.

```python
df.replace("?", np.nan, inplace = True)
df.head(5)
```

```python
df.info()
```

### Learning Points:

* In dealing with missing Data, we can do some methods to resolve:
  1. Droping: This way might cause to remove some columns or rows. This is not a good way. 
  2. Replacing: This is the logical way. There is some methods: 
         a. Reaplace it with Mean
       If you want to do this, be aware that this way just works for <b>numerical</b> types of data. So first of all change intended data type to "int" or "float". (Check the type of the columns data type with <b>df.info()</b>)

```python
mkfloat = df["normalized-losses"].astype("float")
norm_loss_mean = mkfloat.mean()
norm_loss_mean

# norm_loss_mean = df["normalized-losses"].astype("float").mean()
# print("The average of normalized losses is:", norm_loss_mean)
```

```python
df["normalized-losses"].replace(np.nan, norm_loss_mean, inplace = True)
```

```python
df.info()
```

```python
df.head(5)
```
